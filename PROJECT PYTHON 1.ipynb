```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = df['Salary'].mean()
df['Salary'].fillna(x,inplace = True)
df
```

```python
df.dropna ( inplace = True)
df.isnull().sum()
```

```python
df['Height'] = np.random.uniform(150,180,size = len(df))
df
```

```python
df['Team'].value_counts()
```

```python
df['Team'].value_counts()/len(df)*100
```

```python
employees = df.groupby('Position')['Name'].apply(list)
for Position, Names in employees.items():
    print(f"employees in {Position} position:")
    for name in Names:
     print(name)
    print("\n")
```

```python
df['Age Group'] = df['Age'].apply(lambda age:'20-25' if 20 <= age <= 25 else ('26-30' if 26 <= age <= 30 else ('31-35' if 31 <= age <= 35 else '36 and above')))

df
```

```python
df['Age Group'].value_counts()
```

```python
spending_salary = df.groupby(['Team','Position'])['Salary'].sum()
spending_salary.idxmax()
```

```python
corr= df['Age'].corr(df['Salary'])
corr
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation Between Age and Salary')
plt.show()
```
